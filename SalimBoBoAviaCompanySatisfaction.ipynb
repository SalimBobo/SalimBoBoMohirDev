```python
# Kerakli kutubxonalarni import qilish
import matplotlib.pyplot as plt
import seaborn as sns
import lightgbm as lgb
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
# Ma'lumotlarni yuklash funktsiyasi
def load_data():
    train = pd.read_csv('/content/train_dataset.csv',index_col=['id'])
    test = pd.read_csv('/content/test_dataset.csv',index_col=['id'])
    return train, test
```

```python
# Ma'lumotlarni oldindan qayta ishlash funktsiyasi
def preprocess_data(train, test):
    train['test'] = False
    test['test'] = True
    df = pd.concat([train,test],axis=0)
    for c in df.columns[df.dtypes=='object']:
        df[c] = df[c].factorize()[0]
    train = df[df.test==False]
    test = df[df.test==True]
    test = test.drop('satisfaction',axis=1)
    return train, test
```

```python
# Modelni train qilish funktsiyasi
def train_model(X, y):
    model = lgb.LGBMClassifier()
    model.fit(X, y)
    return model
```

```python
# Modelni baholash funktsiyasi
def evaluate_model(model, X, y):
    FOLD = 5
    score = cross_val_score(model, X, y, cv=FOLD, scoring='roc_auc').mean()
    print("Cross-Validation Score :", score)
```

```python
# Submission faylini yaratish funksiyasi
def generate_submission_file(model, test, filename='submission.csv'):
    predictions = model.predict_proba(test)
    sample_sub = pd.read_csv('/content/sample_submission.csv')
    sample_sub['satisfaction'] = 1 - predictions
    sample_sub.to_csv(filename, index=False)

# Kodni ishga tushirish uchun asosiy funktsiya
def main():
    train, test = load_data()
    train, test = preprocess_data(train, test)
    X = train.drop('satisfaction', axis=1)
    y = train.satisfaction
    model = train_model(X, y)
    evaluate_model(model, X, y)
    generate_submission_file(model, test)

# Asosiy funktsiyani ishga tushirish
if __name__ == "__main__":
    main()
```
